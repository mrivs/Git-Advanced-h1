Задача 40: Работать с файлом california_housing_train.csv, который находится в папке sample_data. Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population)
Задача 42: Узнать какая максимальная households в зоне минимального значения population

```python
import pandas
import seaborn
df=pandas.read_csv("california_housing_test.csv")
df.head()
```

```python
data=df[df['population'] < 500]
sr=data["median_house_value"].mean()
print (f"Cредняя стоимость дома, где кол-во людей от 0 до 500: {sr:.2f} $")
```

```python
minpop=500
data=df[df['population'] < minpop]
sr=data["households"].max()
print (f"максимальная households в зоне минимального значения population: {sr:.2f}")
```

```python
seaborn.relplot(x="population", y="households", kind="line", data=df[df["population"]<minpop])
```
